 **Theory** : 1. Role of NumPy in Python Scientific Computing Ecosystem

NumPy (Numerical Python) plays a fundamental role in the Python scientific computing ecosystem as it provides a powerful and efficient way to handle numerical data. It introduces the concept of the N-dimensional array (ndarray), which allows users to store and process large datasets efficiently. NumPy serves as the backbone for many advanced libraries and simplifies complex mathematical computations. It is widely used in fields such as data science, machine learning, engineering, and scientific research, making it an essential tool for numerical computation.

2. Why is NumPy Faster than Pure Python Computations?

NumPy is considered faster than pure Python computations because it is implemented using low-level languages like C and C++, which are optimized for performance. It uses vectorized operations, allowing computations to be performed on entire arrays at once instead of using loops. Additionally, NumPy stores data in contiguous memory locations, which improves memory access speed and overall efficiency. These features significantly reduce execution time and make NumPy more suitable for handling large-scale numerical operations.

3. Types of Mathematical Operations in NumPy

NumPy supports a wide range of mathematical operations that make it highly versatile for numerical computing. It includes basic arithmetic operations such as addition, subtraction, multiplication, and division. It also provides statistical functions like mean, median, and standard deviation for data analysis. Furthermore, NumPy offers linear algebra operations such as matrix multiplication, determinant, and inverse, as well as trigonometric functions like sine, cosine, and tangent. These capabilities allow users to perform both simple and complex calculations efficiently.

4. Indexing and Slicing in NumPy Arrays

Indexing and slicing are important features in NumPy that allow users to access and manipulate array elements efficiently. Indexing is used to retrieve a specific element from an array using its position, while slicing is used to extract a subset of elements from the array. In one-dimensional arrays, slicing can be done by specifying a range of indices, whereas in multi-dimensional arrays, indexing can be performed using row and column positions. These features provide flexibility in handling and modifying data within arrays.

5. Integration of NumPy with Other Libraries

NumPy integrates seamlessly with many other libraries in the Python ecosystem, enhancing its usability and functionality. For example, Pandas uses NumPy arrays as its underlying data structure for efficient data manipulation and analysis. Similarly, Matplotlib relies on NumPy arrays to generate graphs and visualizations from numerical data. In addition, Scikit-learn uses NumPy arrays to handle datasets and perform machine learning tasks. This strong integration makes NumPy a central component of the Python scientific computing ecosystem.



```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Create Student Data (Real-life example)
data = {
    "Name": ["ABHISHEK","SANJANA", "DHIRAJ", "ARPITA", "RAJ"],
    "Maths": [78, 85, 90, 60, 75],
    "Science": [82, 88, 95, 65, 70],
    "English": [75, 80, 85, 70, 72]
}

df = pd.DataFrame(data)

print("Student Data:\n")
print(df)

# 2. NumPy Computation (Average Marks)
marks_array = df[["Maths", "Science", "English"]].values

average_marks = np.mean(marks_array, axis=1)
df["Average"] = average_marks

print("\nStudent Data with Average:\n")
print(df)

# 3. Performance Classification (Simple ML Logic)
df["Result"] = ["Pass" if avg >= 75 else "Fail" for avg in df["Average"]]

print("\nFinal Result:\n")
print(df)

# 4. Visualization using Matplotlib
plt.bar(df["Name"], df["Average"])
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.title("Student Performance Analysis")
plt.show()
```

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Sales Data (Real-life shop example)
data = {
    "Product": ["Paan", "Cold Drink", "Chips", "Chocolate", "Biscuit"],
    "Sales": [200, 350, 150, 180, 220]   # daily sales in rupees
}

df = pd.DataFrame(data)

print("Sales Data:\n")
print(df)

# 2. NumPy Computation
sales_array = df["Sales"].values

total_sales = np.sum(sales_array)
average_sales = np.mean(sales_array)

print("\nTotal Sales:", total_sales)
print("Average Sales:", average_sales)

# 3. Best Selling Product
max_sales = np.max(sales_array)
best_product = df[df["Sales"] == max_sales]["Product"].values[0]

print("\nBest Selling Product:", best_product)

# 4. Performance Category (Simple Prediction Logic)
df["Category"] = ["High" if sale > average_sales else "Low" for sale in df["Sales"]]

print("\nSales with Category:\n")
print(df)

# 5. Visualization
plt.bar(df["Product"], df["Sales"])
plt.xlabel("Products")
plt.ylabel("Sales (Rs)")
plt.title("Daily Sales Analysis")
plt.show()
```
