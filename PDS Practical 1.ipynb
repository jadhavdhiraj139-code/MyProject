**Theory**
1. What are sequence data types in Python? Explain indexing and slicing.

Sequence data types in Python are used to store multiple values in an ordered way. This means every element has a fixed position, and we can access it using its index.

Some common sequence types are strings, lists, tuples, arrays, and range.

In these, indexing is used to access elements. Index always starts from 0, so the first element is at position 0. We can also use negative indexing, where -1 means the last element.

Slicing is used when we want a part of the sequence instead of the whole data. It is written like:
sequence[start:end:step]

This makes it easy to extract or work with specific portions of data.

2. Difference between list and tuple. Why is tuple preferred?

Both list and tuple are used to store multiple values, but the main difference is:

List → can be changed (mutable)
Tuple → cannot be changed (immutable)

Lists use square brackets [ ] and tuples use parentheses ( ).

Tuples are preferred when:

Data should not change
We want better performance
Memory usage should be less

For example, tuples are used for fixed data like coordinates or constant values.

3. Difference between Python arrays and lists. When to use arrays?

Arrays and lists look similar, but they are different in usage.

List → can store different types of data (int, string, etc.)
Array → stores only same type of data

Arrays are more:

Memory efficient
Faster for numerical calculations

Lists are more flexible, so they are used more commonly. But arrays are useful when:

Working with large numerical data
Doing mathematical or scientific operations
4. How do dictionaries work internally? Why are they fast?

Dictionaries store data in key-value pairs.

Internally, Python uses something called a hash table. When we store a value, the key is converted into a unique number (hash), and the value is stored at that position.

Because of this:

Data can be accessed directly using the key
No need to search one by one

This makes dictionaries very fast, usually taking constant time O(1) for searching.

5. Difference between sets and lists. How sets handle duplicates?

Lists and sets both store multiple values, but they work differently.

List
Ordered
Allows duplicates
Set
Unordered
Does not allow duplicates

If we add duplicate values in a set, it automatically removes them and keeps only unique values.

Sets are useful when:

We want only unique data
Removing duplicates
Performing operations like union or intersection

**# String**

```python
# String example
s = "Hello Dhiraj"

print("Original String:", s)

# Indexing
print("First character:", s[0])
print("Last character:", s[-1])

# Slicing
print("Slicing (0:5):", s[0:5])
print("Slicing (6:):", s[6:])
print("Reverse string:", s[::-1])
```

**# List**

```python
# List example
lst = [100, 20, 30, 40, 50]

print("\nOriginal List:", lst)

# Indexing
print("First element:", lst[0])

# Slicing
print("Slice (1:4):", lst[1:4])

# Operations
lst.append(60)
lst.remove(20)

print("Modified List:", lst)
```

**# Array**

```python
import array

# Create array
arr = array.array('i', [1, 2, 3, 4, 5])

print("\nOriginal Array:", arr)

# Indexing
print("First element:", arr[0])

# Slicing
print("Slice (1:4):", arr[1:4])

# Append
arr.append(6)

print("Modified Array:", arr)
```

**# Tuple**

```python
# Tuple example
tup = (100, 200, 300, 400)

print("\nOriginal Tuple:", tup)

# Indexing
print("First element:", tup[0])

# Slicing
print("Slice (1:3):", tup[1:3])
```

**# Dictionary**

```python
# Dictionary example
d = {'name': 'Raj', 'age': 20, 'course': 'Btech'}

print("\nDictionary:", d)

# Access values
print("Name:", d['name'])

# Add new key-value
d['city'] = 'Mumbai'

# Update value
d['age'] = 21

print("Updated Dictionary:", d)
```

**# Sets**

```python
# Set example
s = {1, 2, 3, 4, 4, 5}

print("\nSet:", s)

# Add element
s.add(6)

# Set operations
set2 = {4, 5, 6, 7}

print("Union:", s.union(set2))
print("Intersection:", s.intersection(set2))
```

**# Range**

```python
# Range example
r = range(1, 10)

print("\nRange values:")
for i in r:
    print(i, end=" ")
```
