Implementing Web Scraping in Python with BeautifulSoup.

**Theory :**
1 . What is web scraping? Explain the role of Requests library

Web scraping is a method used to collect data automatically from websites using programs instead of doing it manually. It is helpful when we need large amounts of data like text, images, or links from web pages. In Python, the Requests library is used to connect with websites and get their content. It sends a request to the server and receives the HTML of the webpage in return, just like a browser does. After getting this data, we can use other tools to extract useful information from it.

2.	What is BeautifulSoup? How does it help in parsing HTML?

BeautifulSoup is a Python library that helps in reading and extracting data from HTML or XML files. When we get raw HTML from a website, it is usually complex and difficult to understand. BeautifulSoup makes this easier by organizing the HTML into a structured format, like a tree. This allows us to easily find specific elements such as tags, classes, or IDs and extract the required data. It is very useful even when the HTML code is not properly written.

3.	Difference between web scraping and web crawling

Web scraping and web crawling are often confused, but they are not the same. Web scraping is mainly about collecting specific data from a webpage, like extracting product prices or text. On the other hand, web crawling is about visiting many web pages automatically and collecting information about them. For example, search engines use crawling to discover new pages on the internet. So, crawling is more about exploring websites, while scraping is about extracting useful data from them.

4.	Commonly used methods in BeautifulSoup

While working with BeautifulSoup, there are several methods that make data extraction easier. We can use functions like find() or find_all() to search for specific elements in the HTML. If we want only the text content, we can use get_text(). To access attributes like links or image sources, we can use the get() method. There are also options to navigate through the HTML structure, which helps in moving from one element to another easily. These features make BeautifulSoup very flexible for web scraping tasks.

5.	Ethical and legal considerations in web scraping

While doing web scraping, it is important to follow proper rules and guidelines. Not all websites allow scraping, so we should always check their policies, like the robots.txt file. Sending too many requests in a short time can slow down or damage a website, so it should be avoided. Also, collecting private or copyrighted data without permission can lead to legal problems. So, it is always better to use web scraping in a responsible and ethical way..

Import Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Send Request to website

```python
url = "http://books.toscrape.com/"

response = requests.get(url)

print("Status Code:", response.status_code)
```

Parse HTML content

```python
soup = BeautifulSoup(response.text, 'html.parser')

print(soup.title.text)
```

Extract Book Titles

```python
books = soup.find_all('h3')

titles = []

for book in books:
    title = book.find('a')['title']
    titles.append(title)

titles
```

Extract Prices

```python
prices = soup.find_all('p', class_='price_color')

price_list = []

for price in prices:
    price_list.append(price.text)

price_list
```

Extract Rating

```python
ratings = soup.find_all('p', class_='star-rating')

rating_list = []

for rating in ratings:
    rating_list.append(rating['class'][1])

rating_list
```

Create DataFrame

```python
data = pd.DataFrame({
    'Title': titles,
    'Price': price_list,
    'Rating': rating_list
})

data
```

Save Data to CSV

```python
data.to_csv("books_data.csv", index=False)

print("Data saved successfully!")
```
